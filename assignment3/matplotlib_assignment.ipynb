```python
import matplotlib.pyplot as plt
import numpy as np

# Enable inline plotting if needed
# %matplotlib inline
```

## 1Ô∏è‚É£ Line Plot

```python
# Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

# Simple plot
plt.figure()
plt.plot(x, y, marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()
```

## 2Ô∏è‚É£ Bar Graph

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.figure()
plt.bar(students, marks, color=['blue', 'green', 'purple', 'red'])
plt.title("Marks of Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

## 3Ô∏è‚É£ Pie Chart

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = [0.1, 0, 0, 0]  # Highlight highest revenue

plt.figure()
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%', shadow=True)
plt.title("Revenue Distribution by Region")
plt.show()
```

## 4Ô∏è‚É£ Histogram

```python
# Random integers between 1 and 100
data = np.random.randint(1, 101, size=1000)

plt.figure()
plt.hist(data, bins=20)
plt.title("Frequency Distribution of Random Integers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
