## 1ï¸âƒ£ 
Created a NumPy array with from 1 to 20 using np.arange()

```python
arr_1d = np.arange(1, 21)
print("1D Array (1 to 20):", arr_1d)
```

### 1(a)
Calculated and printed np.sum(), np.mean(), np.median() and np.std()

```python
arr_sum = np.sum(arr_1d)
arr_mean = np.mean(arr_1d)
arr_median = np.median(arr_1d)
arr_std = np.std(arr_1d)

print("Sum:", arr_sum)
print("Mean:", arr_mean)
print("Median:", arr_median)
print("Standard Deviation:", arr_std)
```

### 1(b)
Used np.where to find out the indices of elements greater then 10

```python
indices_gt_10 = np.where(arr_1d > 10)[0]
print("Indices of elements > 10:", indices_gt_10)
```

## 2ï¸âƒ£
Created a 2D NumPy array of shape 4x4 with numbers 1 to 16 using np.arange() and .reshape()

```python
arr_2d = np.arange(1, 17).reshape(4, 4)
```

### 2(a)
Printed the array

```python
print("2D Array (4x4):\n", arr_2d)
```

### 2(b)
Transpose of array using .T

```python
arr_2d_T = arr_2d.T
print("Transpose of the array:\n", arr_2d_T)
```

### 2(c)
Row and column wise sum of array using np.sum()  
axis = 1 ---> row-wise  
axis = 0 ---> column-wise

```python
row_sums = np.sum(arr_2d, axis=1)
col_sums = np.sum(arr_2d, axis=0)

print("Row-wise sums:", row_sums)
print("Column-wise sums:", col_sums)
```

## 3ï¸âƒ£
Creating two 3x3 arrays filled wwith random integers between 1 and 20 using np.random.randint()  
and setting seed for rpeatable results using np.random.seed()

```python
np.random.seed(0)

A = np.random.randint(1, 21, size=(3, 3))
B = np.random.randint(1, 21, size=(3, 3))

print("Array A:\n", A)
print("Array B:\n", B)

```

### 3(a)
Did element wise addition substraction and Multiplication

```python
add_result = A + B
sub_result = A - B
mul_result = A * B

print("Element-wise Addition:\n", add_result)
print("Element-wise Subtraction:\n", sub_result)
print("Element-wise Multiplication:\n", mul_result)
```

### 3(b)
Did dot product of the two arrays using np.dot()

```python
dot_result = np.dot(A, B)
print("Dot Product (A . B):\n", dot_result)
```
