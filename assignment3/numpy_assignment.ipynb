```
# 1. Create a 1D NumPy array with integers from 1 to 20
arr_1d = np.arange(1, 21)
print("1D Array (1 to 20):", arr_1d)

# 1(a). Calculate sum, mean, median, and standard deviation
arr_sum = np.sum(arr_1d)
arr_mean = np.mean(arr_1d)
arr_median = np.median(arr_1d)
arr_std = np.std(arr_1d)

print("Sum:", arr_sum)
print("Mean:", arr_mean)
print("Median:", arr_median)
print("Standard Deviation:", arr_std)

# 1(b). Find indices of elements greater than 10
indices_gt_10 = np.where(arr_1d > 10)[0]
print("Indices of elements > 10:", indices_gt_10)
```

```
# 2. Create a 2D NumPy array of shape 4x4 with numbers from 1 to 16
arr_2d = np.arange(1, 17).reshape(4, 4)
print("2D Array (4x4):\n", arr_2d)

# 2(a). Print the array -> already printed above

# 2(b). Transpose of the array
arr_2d_T = arr_2d.T
print("Transpose of the array:\n", arr_2d_T)

# 2(c). Row-wise and column-wise sums
row_sums = np.sum(arr_2d, axis=1)
col_sums = np.sum(arr_2d, axis=0)

print("Row-wise sums:", row_sums)
print("Column-wise sums:", col_sums)
```

```
# 3. Create two 3x3 arrays filled with random integers between 1 and 20
# Setting a seed for reproducibility (optional)
np.random.seed(0)

A = np.random.randint(1, 21, size=(3, 3))
B = np.random.randint(1, 21, size=(3, 3))

print("Array A:\n", A)
print("Array B:\n", B)

# 3(a). Element-wise addition, subtraction, and multiplication
add_result = A + B
sub_result = A - B
mul_result = A * B

print("Element-wise Addition:\n", add_result)
print("Element-wise Subtraction:\n", sub_result)
print("Element-wise Multiplication:\n", mul_result)

# 3(b). Dot product of the two arrays
dot_result = np.dot(A, B)
print("Dot Product (A Â· B):\n", dot_result)
```

```
# 4. Reshape a 1D array of size 12 into a 3x4 2D array
arr_1d_12 = np.arange(1, 13)
arr_3x4 = arr_1d_12.reshape(3, 4)
print("Original 3x4 array:\n", arr_3x4)

# Slice the first two rows and last two columns
# First two rows -> rows with index 0 and 1 -> :2
# Last two columns -> columns with index 2 and 3 -> -2:
sliced_arr = arr_3x4[:2, -2:]
print("Sliced array (first 2 rows, last 2 columns):\n", sliced_arr)
```
