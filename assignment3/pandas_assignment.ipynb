```python
import pandas as pd
```

## 1Ô∏è‚É£ Create DataFrame and Perform Operations

```python
# Creating the DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
df = pd.DataFrame(data)
df
```

### 1(a) Display first five rows

```python
df.head()
```

### 1(b) Summary statistics of 'Age' and 'Salary'

```python
df[['Age', 'Salary']].describe()
```

### 1(c) Average salary of employees in HR department

```python
df[df['Department'] == 'HR']['Salary'].mean()
```

## 2Ô∏è‚É£ Add Bonus Column

```python
df['Bonus'] = df['Salary'] * 0.10
df
```

## 3Ô∏è‚É£ Filter employees aged between 25 and 30

```python
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
filtered_df
```

## 4Ô∏è‚É£ Group by Department ‚Äî Average Salary

```python
department_avg = df.groupby('Department')['Salary'].mean()
department_avg
```

## 5Ô∏è‚É£ Sort by Salary and Save to CSV

```python
sorted_df = df.sort_values(by='Salary', ascending=True)
sorted_df
```

```python
sorted_df.to_csv('sorted_employees.csv', index=False)
```
